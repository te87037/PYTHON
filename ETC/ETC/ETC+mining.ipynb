```python
import pandas as pd
```

```python
url = "http://tisvcloud.freeway.gov.tw/history/TDCS/M08A/20170529/"
day = "20170529"
hr = ['{0:02d}'.format(x) for x in range(24)]
mins = ['{0:02d}'.format(x) for x in range(60) if x%5==0]


df2 = pd.DataFrame(columns = ["報表產製時間","起點","迄點","車種","流量"])

for hrs in hr:
    for minss in mins:
        data_url = url + str(hrs) +'/' +  "TDCS_M08A_"+ day + "_" + str(hrs) + str(minss) + "00.csv" 
        print (data_url)
        data = pd.read_csv(data_url,header = None,names = ["報表產製時間","起點","迄點","車種","流量"])
        df = pd.DataFrame(data,columns = ["報表產製時間","起點","迄點","車種","流量"])
        df2 = df2.append(df,ignore_index=True)

df.head()

```

```python
Inter = df2['起點'].unique()
```

```python
df2.columns = ["time","start","end","car","flow"]
for i in Inter:
    print("start from {} has ".format(i),end = '')
    print(df2[df2["start"] == i]['flow'].sum())
    
```
